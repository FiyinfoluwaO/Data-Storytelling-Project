# EXPLORATORY DATA ANALYSIS ON MOBILE PHONE DATASET

THE CLEANED DATA OF MOBILE PHONE SCRAPED FROM JUMIA

```python
#To import the cleaned data 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
#Setting up the visual configuration
plt.style.use('ggplot') 
plt.rcParams['font.size'] = 12 
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['font.size'] = 8
plt.rcParams['text.color'] = 'black'
mobile_data=pd.read_csv('Cleaned_Mobile_Phone_Dataset.csv')
```

```python
#To view the description of the dataset
mobile_data.describe(include='all')
```

TO PERFORM DATA VISUALIZATION 

```python
#To plot the brands occurrence
plt.figure(figsize=(12,10))
sns.catplot(y='Brands', data=mobile_data, kind='count')
plt.show()
```

```python
#To visualize the ratings of each brand of mobile phone
plt.figure(figsize=(12,10))
sns.barplot(x='Ratings',y='Brands', data=mobile_data, estimator=np.sum)
plt.title("Brand Ratings")
plt.ylabel("Brands in the Market")
plt.xlabel("Ratings")
plt.show()
```
