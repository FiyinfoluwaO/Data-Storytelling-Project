# DATA CLEANING OF THE MOBILE PHONE RAW DATASET 

```python
#To import the modules
import pandas as pd
import numpy as np
```

```python
#To access the dataset and decode it
phone_data=pd.read_csv('Mobile_Phones_in_Jumia_NIG.csv', encoding = 'Latin-1')
```

```python
#To view the content of the dataset under each headings
phone_data.head()
```

```python
#To determine the number of null values in the dataset
phone_data.isnull().sum()
```

```python
#The forward fill (ffill) method was used to replace the null values
phone_dt=phone_data.fillna(method='ffill')
```

```python
#To check the result of the fill option on the dataset
phone_dt.isnull().sum()
```

```python
#To view the dataset
phone_dt
```

```python
#To drop some columns not needed for the data analysis
phone_dt=phone_dt.drop(['Title_URL','Image','Tag','shipping'],axis=1)
```

```python
#To view the dataset after the removal of some columns
phone_dt
```

```python
#TO EXTRACT SOME VALUES FROM THE STRING COLUMNS 
import re as re
def find_number (text):
    num=re.findall(r'[0-9]+',text)
    return "".join(num)
phone_dt['price']=phone_dt['prc'].apply(lambda x: find_number(x))
phone_dt['Main_Price']=phone_dt['prc_main'].apply(lambda x: find_number(x))
```

```python
#To view the dataset after the EXTRACTION OF VALUES
phone_dt
```

```python
#TO EXTRACT SOME VALUES WITH DECIMAL FROM THE STRING COLUMNS 
phone_dt['Ratings']=phone_dt['ratings'].str.extract('([0-9][,.]*[0-9]*)')
phone_dt['Percent_Price_Reduction_in_%']=phone_dt['percent_reduction'].str.extract('([0-9][,.]*[0-9]*)')
phone_dt['Number_of_Reviewer']=phone_dt['number of reviewer'].str.extract('([0-9][,.]*[0-9]*)')
```

```python
#To view the dataset after the EXTRACTION OF VALUES WITH DECIMAL
phone_dt
```

```python
#TO EXTRACT THE FIRST WORD FROM THE STRINGS IN THE 'TITLE' COLUMNS AS BRAND
phone_dt['Brands']=phone_dt['Title']
phone_dt['Brands']=phone_dt.Title.str.split().str.get(0)
```

```python
#To view the dataset after the update of additional column i.e. BRANDS
phone_dt
```

```python
#To drop some columns not needed for the data analysis AFTER THE CLEANING OF THE DATASET
phone_dt=phone_dt.drop(['prc','prc_main','percent_reduction','ratings','number of reviewer'],axis=1)
```

```python
#To view the dataset after the dropping columns AFTER CLEANING
phone_dt
```

```python
mobile_phone=phone_dt[['Brands','Title','price','Main_Price','Percent_Price_Reduction_in_%','Ratings','Number_of_Reviewer']]
```

```python
mobile_phone
```

```python
mobile_phone.to_csv('Cleaned_Mobile_Phone_Dataset.csv')
```
